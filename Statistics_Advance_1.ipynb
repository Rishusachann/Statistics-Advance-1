**Theory Questions**


---

**Question 1:**
What is a random variable in probability theory?

**Answer:**
A random variable is a quantity that takes different numerical values based on the outcomes of a random experiment. It assigns numbers to events so we can analyze them mathematically.

---

**Question 2:**
What are the types of random variables?

**Answer:**
There are two main types of random variables:

* Discrete random variable: takes countable values (like 0, 1, 2...).
* Continuous random variable: takes any value within a range (like height, weight).

---

**Question 3:**
What is the difference between discrete and continuous distributions?

**Answer:**
Discrete distributions describe probabilities of a discrete random variable, where outcomes are countable. Continuous distributions describe probabilities over continuous ranges, using a probability density function (PDF) because individual exact outcomes have zero probability.

---

**Question 4:**
What are probability distribution functions (PDF)?

**Answer:**
A Probability Distribution Function (PDF) shows how probabilities are distributed over the values of a continuous random variable. The area under the PDF curve in a range gives the probability that the variable falls within that range.

---

**Question 5:**
How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

**Answer:**
A Cumulative Distribution Function (CDF) gives the probability that a random variable is less than or equal to a specific value. A PDF gives the probability density at each value (for continuous) or probability mass (for discrete), while the CDF accumulates these probabilities up to a point.

---

**Question 6:**
What is a discrete uniform distribution?

**Answer:**
A discrete uniform distribution is a probability distribution where all possible outcomes are equally likely. For example, rolling a fair die has outcomes 1-6 each with probability 1/6.

---

**Question 7:**
What are the key properties of a Bernoulli distribution?

**Answer:**
A Bernoulli distribution models a single trial with two possible outcomes: success (1) or failure (0). It has one parameter, p, the probability of success. The mean is p, and variance is p(1-p).

---

**Question 8:**
What is the binomial distribution, and how is it used in probability?

**Answer:**
The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success. It’s used to model events like number of heads in multiple coin tosses.

---

**Question 9:**
What is the Poisson distribution and where is it applied?

**Answer:**
The Poisson distribution gives the probability of a number of events occurring in a fixed interval of time or space, given a constant mean rate of occurrence. It’s used in modeling rare events like phone calls to a call center per hour.

---

**Question 10:**
What is a continuous uniform distribution?

**Answer:**
A continuous uniform distribution has all intervals of the same length within a given range equally likely. For example, choosing a random number between 0 and 1 gives each number the same probability density.

---

**Question 11:**
What are the characteristics of a normal distribution?

**Answer:**
A normal distribution is bell-shaped, symmetric around its mean, with mean = median = mode. It’s defined by mean (µ) and standard deviation (σ), and most data lies within 3σ from the mean.

---

**Question 12:**
What is the standard normal distribution, and why is it important?

**Answer:**
The standard normal distribution is a special case of the normal distribution with mean 0 and standard deviation 1. It is important because many statistical methods use Z-scores, which standardize values using the standard normal.

---

**Question 13:**
What is the Central Limit Theorem (CLT), and why is it critical in statistics?

**Answer:**
The CLT states that the distribution of the sample mean of a large number of independent, identically distributed variables approaches a normal distribution, regardless of the original variable’s distribution. It’s critical because it justifies using normal-based methods for inference.

---

**Question 14:**
How does the Central Limit Theorem relate to the normal distribution?

**Answer:**
The CLT explains why sample means tend to follow a normal distribution when sample size is large, allowing normal distribution-based techniques to be used even if the population distribution is not normal.

---

**Question 15:**
What is the application of Z statistics in hypothesis testing?

**Answer:**
Z statistics are used to test hypotheses about means when the population standard deviation is known or sample size is large. The Z-score helps decide whether to reject the null hypothesis by comparing it with critical values from the standard normal distribution.

---

**Question 16:**
How do you calculate a Z-score, and what does it represent?

**Answer:**
The Z-score is calculated as Z = (X - µ) / σ, where X is the value, µ is the mean, and σ is the standard deviation. It represents how many standard deviations a value is from the mean.

---

**Question 17:**
What are point estimates and interval estimates in statistics?

**Answer:**

* Point estimate: a single value used as an estimate of a population parameter (e.g., sample mean for population mean).
* Interval estimate: a range of values (confidence interval) likely to contain the parameter with a certain level of confidence.

---

**Question 18:**
What is the significance of confidence intervals in statistical analysis?

**Answer:**
Confidence intervals give a range of plausible values for an unknown parameter, providing more information than point estimates and quantifying uncertainty in the estimation process.

---

**Question 19:**
What is the relationship between a Z-score and a confidence interval?

**Answer:**
Z-scores determine the critical values used to build confidence intervals. For example, a 95% confidence interval uses Z-scores of ±1.96 in a standard normal distribution.

---

**Question 20:**
How are Z-scores used to compare different distributions?

**Answer:**
Z-scores standardize values from different distributions, making them comparable by showing how many standard deviations each value is from its own distribution’s mean.

---

**Question 21:**
What are the assumptions for applying the Central Limit Theorem?

**Answer:**

* Samples must be independent.
* Sample size should be large (usually n ≥ 30).
* Variables should come from the same distribution.
* For small samples, population distribution should be roughly normal.

---

**Question 22:**
What is the concept of expected value in a probability distribution?

**Answer:**
Expected value is the long-run average of a random variable over many repetitions, calculated by multiplying each possible outcome by its probability and summing these products.

---

**Question 23:**
How does a probability distribution relate to the expected outcome of a random variable?

**Answer:**
The probability distribution assigns probabilities to possible outcomes of a random variable. The expected outcome (expected value) is calculated from these probabilities, giving a measure of the center of the distribution.

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

values = np.arange(1, 7)
pmf = np.ones_like(values) / len(values)
cdf = np.cumsum(pmf)

plt.step(values, cdf, where='post')
plt.title("CDF of Discrete Uniform Distribution (Die Roll)")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.ylim(0, 1.1)
plt.show()
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)
plt.hist(z_scores, bins=50, density=True)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.show()
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

sample = np.random.normal(52, 5, 100)
pop_mean = 50
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
z_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"Z-statistic={z_stat:.2f}, p-value={p_value:.4f}")
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 200)
mean, std, n = np.mean(data), np.std(data, ddof=1), len(data)
conf_level = 0.95
z_crit = norm.ppf(1 - (1 - conf_level)/2)
margin_error = z_crit * (std / np.sqrt(n))
ci_lower, ci_upper = mean - margin_error, mean + margin_error
print(f"95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
# 21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
from scipy.stats import norm

data = np.random.normal(200, 30, 150)
mean, std, n = np.mean(data), np.std(data, ddof=1), len(data)
z_crit = norm.ppf(0.975)
margin_error = z_crit * (std / np.sqrt(n))
print(f"Mean: {mean:.2f}, 95% Confidence Interval: [{mean - margin_error:.2f}, {mean + margin_error:.2f}]")
```

```python
# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np

data = np.random.normal(0, 1, 1000)
data_range = np.ptp(data)
q75, q25 = np.percentile(data, [75, 25])
iqr = q75 - q25
print(f"Range: {data_range:.2f}, IQR: {iqr:.2f}")
```
